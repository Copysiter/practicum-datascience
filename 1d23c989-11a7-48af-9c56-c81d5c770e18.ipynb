```python
def df_profile(df, name = None):
    display(HTML('<p></p>'))
    if name:
        display(HTML(f'<h3>–ê–Ω–∞–ª–∏–∑ {name}</h3>'))
    display(df.head(5))

    info = df.describe(include='all', datetime_is_numeric=True).T
    info.insert(0, 'dtype', df.dtypes)
    info.insert(5, 'missings', df.isna().sum())
    cat_cols = df.select_dtypes(exclude='number').columns.tolist()
    cat_values = pd.Series([[]] * info.shape[0])
    for col in cat_cols:
        cat_values[col] = df[col].unique()
    info.insert(3, 'values', cat_values)
    display(HTML(f'<h4>–û–ø–∏—Å–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö:</h4>'))
    display(info)

    display(HTML(f'<h4>–û–±—â–∏–µ —Å–≤–µ–¥–µ–Ω–∏—è:</h4>'))
    html = []
    html.append(f'<li>–î–∞—Ç–∞—Å–µ—Ç —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ {df.shape[0]} –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏ {df.shape[1]} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤</li>') 
    if sum(df.isna().sum()) > 0:
        missing = df.isna().sum()[df.isna().sum() > 0]
        missing_cols = [f'<i><b>{i}</b> ({missing[i]})</i>' for i in missing.index]
        html.append('<li>–î–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –ø—Ä–æ–ø—É—Å–∫–µ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö: {}</li>'.format(', '.join(missing_cols)))
    else:
        html.append('<li>–ü—Ä—É–ø—É—Å–∫–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –æ—Ç—Å—É—Ç—Å–≤—Ç–≤—É—é—Ç</li>')
    if df.duplicated().sum() > 0:
        html.append('<li>–ù–∞–π–¥–∏–Ω–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {}</li>'.format(df.duplicated().sum()))
    else:
        html.append('<li>–î—É–±–ª–∏–∫–∞—Ç—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã</li>')
        
    display(HTML('<ul>{}</ul>'.format(''.join(html))))
```

### –ê–Ω–∞–ª–∏–∑ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
def continuous_number_summary(df, column, target, bins=30, title='', xlabel=''):
    
    #display(HTML(f'<h3>{column}:</h3>'))
    
    display(HTML(f'<p><h3>{title}</h3></p>'))
    
    quantiles = [0.01, 0.25, 0.50, 0.75, 0.99]
    summary = df.groupby(target)[column].describe(percentiles=quantiles).reset_index(drop=True)
    summary.index = summary.index.map({0: '–ê–∫—Ç–∏–≤–Ω—ã–µ –∫–ª–∏–µ–Ω—Ç—ã', 1: '–£—à–µ–¥—à–∏–µ –∫–ª–∏–µ–Ω—Ç—ã'})
    summary.columns=['count', 'mean', 'std', 'min', '1%', '25%', '50%', '75%', '99%', 'max']
    display(summary)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    #fig.suptitle(f'\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –≥–æ–¥—É –Ω–∞—á–∞–ª–∞ –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —É—Å–ª—É–≥–∞–º–∏ –∫–æ–º–ø–∞–Ω–∏–∏\n', fontsize=18)
    
    df.plot(kind='hist', y=column, ec='black', bins=bins,
            linewidth=1, alpha=0.25, label='–í—Å–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[0], color='grey')
    df[df[target] == 1].plot(kind='hist', y=column, ec='black', bins=bins, linewidth=1, alpha=0.75,
                             label='–£—à–µ–¥—à–∏–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[0],grid=True, legend=True, color='r')
    
    df[column].plot(kind='kde', color= 'grey', label= '–í—Å–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[1], legend=True)
    df[df[target] == 1][column].plot(kind='kde', color= 'r', label= '–£—à–µ–¥—à–∏–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[1], legend=True)
    #sns.kdeplot(df[column], color= 'grey', label= '–ê–∫—Ç–∏–≤–Ω—ã–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[1])
    #sns.kdeplot(df[df[target] == 1][column], color= 'red', label= '–£—à–µ–¥—à–∏–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[1])
    
    #plt.xticks(rotation = 0)
    
    axes[0].set_title('\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π\n', fontsize=14)
    axes[0].set_xlabel(f'\n{xlabel}')
    axes[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π\n')
    
    axes[1].set_title(f'\n–ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è\n', fontsize=14)
    axes[1].set_xlabel(f'\n{xlabel}')
    axes[1].set_ylabel('\n')
    
    plt.tight_layout(pad=2)
    plt.show()
```

```python
continuous_number_summary(df, 'duration_months', 'churn', bins=df['duration_months'].nunique(),
                          title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —É—Å–ª—É–≥–∞–º–∏ –∫–æ–º–ø–∞–Ω–∏–∏:',
                          xlabel='–ü–µ—Ä–∏–æ–¥ –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —É—Å–ª—É–≥–∞–º–∏')

continuous_number_summary(df, 'monthly_charges', 'churn', bins=30,
                          title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –µ–∂–µ–º–µ—Å—è—á–Ω–æ–≥–æ –ø–ª–∞—Ç–µ–∂–∞:',
                          xlabel='–†–∞–∑–º–µ—Ä –µ–∂–µ–º–µ—Å—è—á–Ω–æ–≥–æ –ø–ª–∞—Ç–µ–∂–∞')

continuous_number_summary(df, 'total_charges', 'churn', bins=30,
                          title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ–±—â–∏—Ö —Ç—Ä–∞—Ç –Ω–∞ —É—Å–ª—É–≥–∏ –∫–æ–º–ø–∞–Ω–∏–∏:',
                          xlabel='–û–±—â–∏–µ —Ç—Ä–∞—Ç—ã –Ω–∞ —É—Å–ª—É–≥–∏')
```

### –ê–Ω–∞–ª–∏–∑ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
def discrete_number_summary(df, column, target, title=''):
    
    #display(HTML(f'<h3>{column}:</h3>'))
    
    display(HTML(f'<p><h3>{title}</h3></p>'))
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    #fig.suptitle(f'\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –≥–æ–¥—É –Ω–∞—á–∞–ª–∞ –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —É—Å–ª—É–≥–∞–º–∏ –∫–æ–º–ø–∞–Ω–∏–∏\n', fontsize=18)
    
    df.groupby(column)[column].count().plot(kind='bar', ec='black',
            linewidth=1, alpha=0.25, label='–í—Å–µ –∫–ª–∏–µ–Ω—Ç—ã', legend=True, ax=axes[0], width=0.8, color='grey')
    df[df[target] == 1].groupby(column)[column].count().plot(kind='bar', ec='black', linewidth=1, alpha=0.75,
                             label='–£—à–µ–¥—à–∏–µ –∫–ª–∏–µ–Ω—Ç—ã', ax=axes[0], grid=True, legend=True, width=0.8, color='r')
    
    (df[df[target] == 1][column].value_counts() / df[column].value_counts()).sort_index().plot(kind='line', 
                              linewidth=3, alpha=1, ax=axes[1], label='–î–æ–ª—è —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤', legend=True, color='r')
    
    axes[0].set_title(f'\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π\n', fontsize=14)
    axes[0].set_ylabel('')
    axes[1].set_title(f'\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ–ª–∏ —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤\n', fontsize=14)
    axes[1].set_ylabel('')
    
    axes[0].set_xticks(axes[0].get_xticks())
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
    plt.tight_layout(pad=2)
    plt.show()
    
    print()
```

```python
def model_report(grid_search, X, y, label=None):
    
    model = grid_search.best_estimator_
    
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)
    y_proba_one = y_proba[:, 1]
    
    roc_auc = roc_auc_score(y, y_proba_one)
    accuracy = accuracy_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    average_precision = average_precision_score(y, y_proba_one)
    
    print()
    if label:
        results.loc[label, 'AUC-ROC'] = roc_auc
        results.loc[label, 'Accuracy'] = accuracy
        results.loc[label, 'Precision'] = precision
        results.loc[label, 'Recall'] = recall
        results.loc[label, 'F1'] = f1
        display(HTML(
            '<p><b>–õ—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC-ROC:</b> {:.2f}<p><b>–õ—É—á—à–∏–µ –ü–∞—Ä–∞–º–µ—Ç—Ä—ã</b>: {}'.format(
                roc_auc,
                grid_search.best_params_
            )
        ))
    else:
        display(HTML(
            '<p><b>–ó–Ω–∞—á–µ–Ω–∏–µ AUC-ROC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:</b> {:.2f}</p>'.format(roc_auc)
        ))
    print()
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6)) 
    
    fpr, tpr, thresholds = roc_curve(y, y_proba_one)
    lw = 2
    axes[0].plot(fpr, tpr, lw=lw, color='grey')
    axes[0].plot([0, 1], [0, 1], color='r')
    axes[0].set_xlim([-0.02, 1])
    axes[0].set_ylim([0.0, 1.02])
    axes[0].set_xlabel('\nFalse Positive Rate')
    axes[0].set_ylabel('True Positive Rate\n')
    axes[0].text(0.6, 0.2,
                'ROC-AUC = {:.2f}\n\nAccuracy = {:.2f}\n\nPrecision = {:.2f}\n\nRecall = {:.2f}\n\nF1 = {:.2f}'
                 .format(roc_auc, accuracy, precision, recall, f1))
    
    cm = confusion_matrix(y, y_pred)
    names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
    counts = [value for value in cm.flatten()]
    pcts = ['{0:.2%}'.format(value) for value in cm.flatten() / np.sum(cm)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(names, counts, pcts)]
    labels = np.asarray(labels).reshape(2, 2)
    sns.heatmap(cm, annot=labels, cmap='Reds', fmt ='', ax=axes[1]) 
    axes[1].set_xlabel('\nPredict')
    axes[1].set_ylabel('True value\n')
    
    
    '''
    # –±–ª–æ–∫ –¥–ª—è PR-–∫—Ä–∏–≤–æ–π
    precision_plt, recall_plt, thresholds = precision_recall_curve(y, y_proba_one)
    axes[2].step(recall_plt, precision_plt, where='post')
    axes[2].set_xlabel('Recall(Positive label: 1)')
    axes[2].set_ylabel('Precision(Positive label: 1)')
    axes[2].set_xlim([-0.02, 1])
    axes[2].set_ylim([0.0, 1.02])
    axes[2].text(0.2, 0.3, 'Average precision = ' + str(average_precision)[:5])
    '''
    
    plt.tight_layout()
    plt.show()
    
    report = classification_report(
        y, y_pred,
        target_names=['Non-churned', 'Churned'], output_dict=True
    )
    display(pd.DataFrame(report).T)
```

–†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ (–Ω–µ–Ω—É–∂–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —É–¥–∞–ª–∏–º):

```python
X = df.drop(
    columns=[
        'churn', 'begin_year', 'end_year', 'duration_months', 'total_charges', 'gender',
        'multiple_lines', 'streaming_tv', 'streaming_movies',
    ]
)
y = df['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=SEED)
```

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –≤–µ—Ä–Ω–æ, –ø–ª—é—Å –∑–∞ stratify

–í—ã—á–∏—Å–ª–∏–º –≤–µ—Å–∞ –∫–ª–∞—Å—Å–æ–≤ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞:

```python
CLASS_WEIGHTS = (1, round(y_train.value_counts()[0] / y_train.value_counts()[1], 1))
```

–†–∞–∑–¥–µ–ª—è–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ —á–∏—Å–ª–æ–≤—ã–µ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ:

```python
NUM_COLUMNS = X_train.select_dtypes(include='number').columns.tolist()
CAT_COLUMNS = X_train.select_dtypes(exclude='number').columns.tolist()
```

–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ä–∞–∑–¥–µ–ª–∏–º –Ω–∞ –¥–≤–µ –≥—Ä—É–ø–ø—ã –¥–ª—è —Ä–∞–∑–Ω–æ–≥–æ —Ç–∏–ø–∞ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:

```python
OHE_COLUMNS = []
OE_COLUMNS = []

for col in CAT_COLUMNS:
    if df[col].nunique() > 2:
        OHE_COLUMNS.append(col)
    else:
        OE_COLUMNS.append(col)
```

–ò–∑–º–µ–Ω–∏–º —Ç–∏–ø –≤—Å–µ—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ 'category':

```python
X_train.loc[:, CAT_COLUMNS] = X_train[CAT_COLUMNS].astype('category')
X_test.loc[:, CAT_COLUMNS] = X_test[CAT_COLUMNS].astype('category')
```

–°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ –æ–±—É—á–∞–µ–º—ã–º –º–æ–¥–µ–ª—è–º:

```python
results = pd.DataFrame(
    [[0] * 6] * 3,
    index=['LogisticRegression', 'RandomForestClassifier', 'CatBoostClassifier'],
    columns=['AUC-ROC', 'Accuracy', 'Precision', 'Recall', 'F1', 'Grid Search Time'])

pd.set_option('display.float_format', '{: .2f}'.format)
```

### LogisticRegression

–î–ª—è –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å GridSearchCV —Å –∏—Å–ø–æ—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–∞–π–ø–ª–∞–π–Ω–∞ –¥–ª—è –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ö –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º –ø—Ä–∏–º–µ–Ω–∏–º OneHoteEncoder (–µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –±–æ–ª–µ 2) –∏–ª–∏ OrdinalEncoder (–¥–ª—è –±–∏–Ω–∞—Ä–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤). –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–µ–º

```python
%%time
start = time.time()

transformer_logictic_regression = make_column_transformer(
    (
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False, dtype='int'),
        OHE_COLUMNS
    ),
    (
        OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1),
        OE_COLUMNS
    ),
    (
        StandardScaler(), 
        NUM_COLUMNS
    ),
    remainder='passthrough'
)

model_logictic_regression = LogisticRegression(solver='liblinear')

pipeline_logictic_regression = Pipeline([
    ('transformer' , transformer_logictic_regression),
    ('model', model_logictic_regression)
])

param_logictic_regression = {
    'model__penalty': ['l1', 'l2'],
    'model__C': np.logspace(-4, 4, 7)
}

state = np.random.RandomState(SEED)
cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=state)

grid_search_logictic_regression = GridSearchCV(
    pipeline_logictic_regression, 
    param_grid=param_logictic_regression, 
    scoring='roc_auc', cv=cv, n_jobs=-1
)
grid_search_logictic_regression.fit(X_train, y_train)

results.loc['LogisticRegression', 'Grid Search Time'] = time.time() - start
```

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –ø–æ –æ–±—É—á–µ–Ω–∏—é –≤–µ—Ä–Ω–æ

```python
model_report(grid_search_logictic_regression, X_train, y_train, 'LogisticRegression')
```

### RandomForestClassifier

–î–ª—è –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å RandomizedSearchCV (–¥–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±—É—á–µ–Ω–∏—è) —Å –∏—Å–ø–æ—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–∞–π–ø–ª–∞–π–Ω–∞ –¥–ª—è –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ü–æ—Å–∫–æ–ª—å–∫—É OneHoteEncoder –Ω–µ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –º–æ–¥–µ–ª–µ–π, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã—Ö –Ω–∞ –¥–µ—Ä–µ–≤—å—è—Ö, –∫–æ –≤—Å–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º –ø—Ä–∏–º–µ–Ω–∏–º OrdinalEncoder. –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–µ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º.

```python
%%time
start = time.time()

transformer_random_forest = make_column_transformer(
    (
        OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1),
        CAT_COLUMNS
    ),
    remainder='passthrough'
)

model_random_forest = RandomForestClassifier(random_state=SEED)

pipeline_random_forest = Pipeline([
    ('transformer' , transformer_random_forest),
    ('model', model_random_forest)
])

param_random_forest = {
    'model__n_estimators': np.arange(100, 1001, 10),
    'model__max_depth': np.arange(2, 16, 1),
    #'model__max_features' : np.arange(4, len(X_train.columns) + 1, 1)
}

state = np.random.RandomState(SEED)
cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=state)

grid_search_random_forest = RandomizedSearchCV(
    pipeline_random_forest, 
    param_distributions=param_random_forest, 
    scoring='roc_auc', cv=cv, n_iter=100, n_jobs=-1
)
grid_search_random_forest.fit(X_train, y_train)

results.loc['RandomForestClassifier', 'Grid Search Time'] = time.time() - start
```

```python
model_report(grid_search_random_forest, X_train, y_train, 'RandomForestClassifier')
```

### CatBoostClassifier

–î–ª—è –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å RandomizedSearchCV (–¥–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±—É—á–µ–Ω–∏—è) —Å –∏—Å–ø–æ—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–∞–π–ø–ª–∞–π–Ω–∞. –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—å –Ω–µ –±—É–¥–µ–º, –ø–æ—Å–∫–æ–ª—å–∫—É –≤ –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–π –º–æ–¥–µ–ª–∏ –µ—Å—Ç—å –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ –º–µ—Ç –æ–¥—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. 

```python
%%time
start = time.time()

model_catboost = CatBoostClassifier(
    silent=True, class_weights=CLASS_WEIGHTS,
    eval_metric='AUC', random_seed=SEED
)

pipeline_catboost = Pipeline([
    ('model', model_catboost)
])

param_catboost = {
    'model__iterations': np.arange(100, 1001, 10),
    'model__learning_rate': np.arange(0.1, 0.5, 0.1),
    'model__l2_leaf_reg': np.arange(1, 6, 1), 
    'model__depth': np.arange(2, 6, 1)
}

state = np.random.RandomState(SEED)
cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=state)

grid_search_catboost = RandomizedSearchCV(
    pipeline_catboost, 
    param_distributions=param_catboost, 
    scoring='roc_auc', cv=cv, n_iter=100, n_jobs=-1
)
grid_search_catboost.fit(X_train, y_train, model__cat_features=CAT_COLUMNS)

results.loc['CatBoostClassifier', 'Grid Search Time'] = time.time() - start
```

```python
model_report(grid_search_catboost, X_train, y_train, 'CatBoostClassifier')
```
